```python
import numpy as np
import gurobipy as gp
from gurobipy import GRB

demand = [750, 100, 50, 100, 400, 1000]
setupcost = 400
holdingcost = 2

m = gp.Model("Wagner-Whitin")

# Create variables
lotsize = {}
inventory ={}
setup = {}
for t in range(6):
   lotsize[t] = m.addVar() #use default setting of arguments
   inventory[t] = m.addVar(obj=holdingcost) #objective function is set here
   setup[t] = m.addVar(vtype=GRB.BINARY, obj=setupcost) #objective function is here

#6 constraints for each type of constraint => 6 periods

# inventory balance
m.addConstr(inventory[0] == lotsize[0]-demand[0])
for t in range(1,6):
   m.addConstr(inventory[t-1]+lotsize[t]-demand[t] == inventory[t])
    
# Logic constraints
for t in range(6):
   m.addConstr(lotsize[t] <= 1000*setup[t]) #1000 ~ big-M, may choose max demand 

m.optimize() #default is minimize, if maximize need to change

print("Objective: "+str(m.objVal))
for v in m.getVars():
  print(v.x)
```
