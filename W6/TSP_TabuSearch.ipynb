```python
# Travelling salesman - 120 cities

import numpy as np
import pandas as pd
import math
import random

#coordinates = pd.read_excel("gr120.xlsx", sheet_name="Coordinates")
#co=coordinates.values

cities = 120

# Compute distance matrix

#distance=np.zeros((cities,cities))

#for i in range(cities):
#    for j in range(cities):
#        if (i==j):
#           distance[i,j]=10000
#        else: distance[i,j]=math.sqrt((co[i,0]-co[j,0])*(co[i,0]-co[j,0])+(co[i,1]-co[j,1])*(co[i,1]-co[j,1]))

di = pd.read_excel("gr120.xlsx",sheet_name="DistanceMatrix")
distance=di.values

for i in range(cities):
    distance[i,i] = 10000

# nearest neighbor    

position = 0
tour = [0]
length = 0
for i in range(cities-1):
  nn = 0
  nd = 10000
  for j in range(cities):  
    if (j not in tour) and (distance[position,j]<nd):
      nd = distance[position,j]
      nn = j
  tour.append(nn)
  length = length + nd
  position = nn
tour.append(0)
length = length + distance[position,0]

#random.seed(987654321)
best = length
besttour = tour
tabu = []
for k in range(200):
  maximprove = -length
  for i in range(cities-2):
    for j in range(i+2,cities):
      if [i,j] not in tabu:  # only consider options not already in tabu list
        delta = distance[tour[i],tour[i+1]]+distance[tour[j],tour[j+1]]-distance[tour[i],tour[j]]-distance[tour[i+1],tour[j+1]]
        for k in range(j-i-1):
          delta=delta+distance[tour[j-k-1],tour[j-k]]-distance[tour[j-k],tour[j-k-1]]    
        if delta > maximprove:
          best_i = i
          best_j = j
          maximprove = delta
  next_tour = []
  for k in range(best_i+1):
    next_tour.append(tour[k])
  for k in range(best_j-best_i):
    next_tour.append(tour[best_j-k])
  for k in range(cities-best_j):
    next_tour.append(tour[best_j+k+1]) 
  length = length - maximprove
  tour = next_tour
  tabu.append([best_i,best_j])
# tabu list must be long enough to escape from local optima
  if len(tabu) > 20: #maintain length of tabu list less or equal to 20 => if over 20, add new also delete the oldest
     tabu.pop(0)
  if length < best:
    best = length
    besttour=tour
    print(best)
print("Terminate")

```
