```python
# Benders decomposition for warehouse location problem

import gurobipy as gp
from gurobipy import *
import numpy as np

locations = 3
customers = 5

fixcost = [2, 3, 3]

cost = [[2,3,4,5,7],[4,3,1,2,6],[5,4,2,1,3]]

# Total WLP as MILP

m = gp.Model("Benders WLP")

y = {}
ship = {}
for i in range(locations):
    y[i] = m.addVar(obj=fixcost[i],vtype=GRB.BINARY)
for i in range(locations):
  for j in range(customers):
    ship[i,j] = m.addVar(obj=cost[i][j])
    
for j in range(customers):
    m.addConstr(sum(ship[i,j] for i in range(locations)) >= 1)
    
for i in range(locations):
    for j in range(customers):
        m.addConstr(ship[i,j] <= y[i])
    
m.optimize()

print(m.objval)

# Subproblem for given y

bsp = gp.Model("Benders Subproblem")

y = [1,0,0]
x = {}
for i in range(locations):
  for j in range(customers):
    x[i,j] = bsp.addVar(obj=cost[i][j])
    
for j in range(customers):
    bsp.addConstr(sum(x[i,j] for i in range(locations)) >= 1)
    
for i in range(locations):
    for j in range(customers):
        bsp.addConstr(x[i,j] <= y[i])
    
bsp.optimize()

# Getting the dual variables

print(bsp.objval)

v = np.zeros(customers)
w = np.zeros((locations,customers))
i=0
for c in bsp.getConstrs():
    print(c.getAttr("Pi"))
    if i < customers:
      v[i] = c.getAttr("Pi")
    else: 
      lo = (i-customers) // customers
      cu = (i-customers) % customers
      w[lo,cu] = c.getAttr("Pi")
    i=i+1

for i in range(customers):
  print(v[i])
for i in range(locations):
   for j in range(customers): 
      print(w[i,j])
```
