```python
# Travelling salesman - 120 cities

import numpy as np
import pandas as pd
import math
import random

#coordinates = pd.read_excel("gr120.xlsx", sheet_name="Coordinates")
#co=coordinates.values

cities = 120

# Compute distance matrix

#distance=np.zeros((cities,cities))

#for i in range(cities):
#    for j in range(cities):
#        if (i==j):
#           distance[i,j]=10000
#        else: distance[i,j]=math.sqrt((co[i,0]-co[j,0])*(co[i,0]-co[j,0])+(co[i,1]-co[j,1])*(co[i,1]-co[j,1]))

di = pd.read_excel("gr120.xlsx",sheet_name="DistanceMatrix")
distance=di.values

for i in range(cities):
    distance[i,i] = 10000

# nearest neighbor    

position = 0
tour = [0]
length = 0
for i in range(cities-1):
  nn = 0
  nd = 10000
  for j in range(cities):  
    if (j not in tour) and (distance[position,j]<nd):
      nd = distance[position,j]
      nn = j
  tour.append(nn)
  length = length + nd
  position = nn
tour.append(0)
length = length + distance[position,0]

print('start tour', tour)
print('start length', length)

random.seed(987654321)
best = length
besttour = tour
temperature = 0.01*length
while temperature > 1: 
  print(best, temperature)
  for k in range(10000):
    i = random.randrange(cities-2)
    j = random.randrange(i+2,cities)
    delta = distance[tour[i],tour[i+1]]+distance[tour[j],tour[j+1]]-distance[tour[i],tour[j]]-distance[tour[i+1],tour[j+1]]
    for k in range(j-i-1):
      delta=delta+distance[tour[j-k-1],tour[j-k]]-distance[tour[j-k],tour[j-k-1]]    
    if delta > 0 or (random.random() < math.exp(delta/temperature)): 
      
        #if search for all nodes (steepest descent) => may get back to local optimal
        #randomly => search for more redundant nodes and may also get back to local optimal => tabu search
      next_tour = []
      for k in range(i+1):
        next_tour.append(tour[k])
      for k in range(j-i):
        next_tour.append(tour[j-k])
      for k in range(cities-j):
        next_tour.append(tour[j+k+1]) 
      print('****', length, delta)
      length = length - delta
      tour = next_tour
    if length < best:
      best = length
      besttour=tour
      print(best, '   ', temperature)
  temperature = 0.95*temperature                     
print("Terminate")
print(temperature)
print(besttour)

```
